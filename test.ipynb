```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from fuzzyDB import Fuzzy
```

```python
fuzzy = Fuzzy()
fuzzy.domain_variabel['Harga'] =  np.arange(0, 201000, 500)
fuzzy.domain_variabel['Fasilitas'] = np.arange(0, 26, 0.5)
fuzzy.domain_variabel['Objek_Wisata_Berdekatan'] = np.arange(0, 26, 0.5)
fuzzy.domain_variabel['Jarak_dari_Pusat_Kota'] = np.arange(0, 75, 0.1)

fuzzy.himpunan_fuzzy['Harga'] = {
    'Murah': [0, 0, 15000, 20000],
    'Menengah': [15000, 37500, 60000],
    'Mahal': [50000, 60000, 200000, 200000]
                                }
fuzzy.himpunan_fuzzy['Fasilitas'] = {
    'Sedikit': [0, 0, 3, 5],
    'Sedang': [3, 5, 8],
    'Banyak': [7, 10, 25, 25]
}

fuzzy.himpunan_fuzzy['Objek_Wisata_Berdekatan'] = {
    'Sedikit': [0, 0, 3, 5],
    'Sedang': [3, 5, 8],
    'Banyak': [7, 10, 25, 25]
}

fuzzy.himpunan_fuzzy['Jarak_dari_Pusat_Kota'] = {
    'Dekat': [0, 0, 5, 10],
    'Menengah': [6, 15, 24],
    'Jauh': [20, 30, 40],
    'Sangat_Jauh': [36, 45, 75, 75]
}
```

```python
fuzzy.domain_variabel
```

```python
for k,v in fuzzy.domain_variabel.items():
    plt.title(k.replace('_', ' ').strip())
    for i, j in fuzzy.himpunan_fuzzy[k].items():
        if len(j) == 4:
            plt.plot(j, [fuzzy.trapmf(x, *j) for x in j], label=i.replace('_', ' ').strip())
        else:
            plt.plot(j, [fuzzy.trimf(x, *j) for x in j], label=i.replace('_', ' ').strip())
    plt.legend()
    plt.show()
```

```python
df = pd.read_csv('dataset/pariwisata.csv')
df
```

```python
selection_df = df[['Harga', 'Fasilitas', 'Objek_Wisata_Berdekatan', 'Jarak_dari_Pusat_Kota']]
for i in selection_df.columns:
    for k, v in fuzzy.himpunan_fuzzy[i].items():
        if len(v) == 4:
            df[i + ' ' + k.replace('_', ' ').strip()] = [fuzzy.trapmf(x, *v) for x in df[i].to_list()]
        else:
            df[i + ' ' + k.replace('_', ' ').strip()] = [fuzzy.trimf(x, *v) for x in df[i].to_list()]
```
